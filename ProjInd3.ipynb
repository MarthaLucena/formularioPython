**Importando as bibliotecas:**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

**Carregar o arquivo CSV:**

```python
# Carregar o arquivo CSV
df = pd.read_csv('/content/data.csv')
```

**Visualização dos 5 primeiros itens do arquivo:**

```python
df.head()
```

**Geração de dois exemplos de gráfico:**


1.   *Gráfico de Bugs resolvidos por funcionário;*
2.   *Gráfico de Tarefas Concluídas por funcionário.*



```python
# Gráfico de Bugs Resolvidos
plt.figure(figsize=(8, 4))
plt.bar(df['Nome'], df['Bugs'], color='skyblue')
plt.title('Bugs Resolvidos por Funcionário')
plt.xlabel('Funcionário')
plt.ylabel('Bugs Resolvidos')
plt.xticks(rotation=45)
plt.show()

# Gráfico de Tarefas Concluídas
plt.figure(figsize=(8, 4))
plt.bar(df['Nome'], df['Tarefas'], color='lightgreen')
plt.title('Tarefas Concluídas por Funcionário')
plt.xlabel('Funcionário')
plt.ylabel('Tarefas Concluídas')
plt.xticks(rotation=45)
plt.show()
```
